```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from xgboost import XGBClassifier
```

```python
X,y=load_breast_cancer(return_X_y=True)
X=pd.DataFrame(X)
X.columns=X.columns.astype(str)
```

```python
X, y = load_breast_cancer(return_X_y=True)
X=pd.DataFrame(X)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.model_selection import GridSearchCV
params = {
        "n_estimators": [50,100,150,200],
        "max_depth" : [3, 4, 5, 7],
        "learning_rate": [0.1, 0.2, 0.3],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        }
xgb = XGBClassifier()
```

```python
grid_search_cv=GridSearchCV(xgb,param_grid=params,
                                   scoring='accuracy',
                                   n_jobs=-1,
                                   cv=3,
                                   verbose=2)
```

```python
grid_search_cv.fit(X_train,y_train)
```

```python
grid_search_cv.best_estimator_
```

```python
xgb=grid_search_cv.best_estimator_
```

```python
xgb.fit(X_train,y_train)
```

```python
xgb.score(X_test,y_test)
```

```python
print("Model acc",xgb.score(X_test, y_test))
```

## Light GBM

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
```

```python
data = load_wine()
X = data.data
y = data.target
```

```python
X=pd.DataFrame(X)
X
```

```python
y
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
from lightgbm import LGBMClassifier as lgbc
from sklearn.model_selection import GridSearchCV
params={
    'learning_rate': [0.1, 0.3, 0.5],
    'boosting_type' : ['gbdt'],
    'objective' : ['multiclass'],
    'max_depth' : [5,6,7,8],
    'colsample_bytree' : [0.5,0.7],
    'subsample' : [0.5,0.7],
    'metric':['multi_error']
}
```

```python
model=lgbc()
random_cv = GridSearchCV(model,params,verbose=3,cv=3,n_jobs = -1)
```

```python
random_cv.fit(X_train,y_train)
res=random_cv.best_estimator_
```

```python
res
```

```python
model=res
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
print("Accuracy :", model.score(X_test, y_test))
```
